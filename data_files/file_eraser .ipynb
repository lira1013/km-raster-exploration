```python
import os
import stat

# ç¬¬ä¸€æ­¥ï¼šæ£€æŸ¥ç›®æ ‡æ–‡ä»¶ï¼ˆåŒæ—¶æŸ¥txtå’Œpngï¼‰
def check_target_files(base_dir):
    # å®šä¹‰è¦åˆ é™¤çš„æ–‡ä»¶åˆ—è¡¨
    target_files = ["yolo_labels.txt", "shadow_verify.png"]
    found_files = []
    
    for root, dirs, files in os.walk(base_dir):
        for file in files:
            if file in target_files:
                file_path = os.path.join(root, file)
                found_files.append(file_path)
    
    if found_files:
        print(f"âœ… æ‰¾åˆ° {len(found_files)} ä¸ªç›®æ ‡æ–‡ä»¶ï¼š")
        for f in found_files:
            print(f"   - {f}")
    else:
        print(f"âŒ ä»æœªæ‰¾åˆ°ä»»ä½•ç›®æ ‡æ–‡ä»¶ï¼è¯·æ£€æŸ¥ç›®å½•ï¼š{base_dir}")
    return found_files

# ç¬¬äºŒæ­¥ï¼šå¼ºåˆ¶åˆ é™¤ï¼ˆåŒæ—¶åˆ txtå’Œpngï¼‰
def force_clean_target_files(base_dir):
    # å…ˆæ£€æŸ¥æ–‡ä»¶
    found_files = check_target_files(base_dir)
    if not found_files:
        return
    
    confirm = input("\nâš ï¸ ç¡®è®¤åˆ é™¤ä»¥ä¸Šæ‰€æœ‰æ–‡ä»¶ï¼Ÿ(y/n)ï¼š")
    if confirm.lower() != 'y':
        print("âŒ å–æ¶ˆåˆ é™¤ã€‚")
        return
    
    delete_count = 0
    failed_files = []
    target_files = ["yolo_labels.txt", "shadow_verify.png"]
    
    for root, dirs, files in os.walk(base_dir):
        for file in files:
            if file in target_files:
                file_path = os.path.abspath(os.path.join(root, file))
                try:
                    # è§£å†³æƒé™é—®é¢˜ï¼šæ·»åŠ å¯å†™æƒé™
                    os.chmod(file_path, stat.S_IWRITE)
                    # å¼ºåˆ¶åˆ é™¤æ–‡ä»¶
                    os.remove(file_path)
                    delete_count += 1
                    print(f"âœ… å·²åˆ é™¤ï¼š{file_path}")
                except Exception as e:
                    print(f"âŒ åˆ é™¤å¤±è´¥ {file_path}ï¼š{e}")
                    failed_files.append(file_path)
    
    print(f"\nğŸ§¹ æ¸…ç†å®Œæˆï¼")
    print(f"   âœ… æˆåŠŸåˆ é™¤ï¼š{delete_count} ä¸ªæ–‡ä»¶")
    print(f"   âŒ å¤±è´¥åˆ é™¤ï¼š{len(failed_files)} ä¸ªæ–‡ä»¶")
    if failed_files:
        print(f"   å¤±è´¥åˆ—è¡¨ï¼š{failed_files}")

# æ‰§è¡Œï¼šæ›¿æ¢ä¸ºä½ ç¡®è®¤çš„æ­£ç¡®ç»å¯¹è·¯å¾„
CORRECT_BASE_DIR = "/mnt/ONCOBOX/geneplus/workspace/chenly/KM_DATA/cbioprotal_download_km"
force_clean_target_files(CORRECT_BASE_DIR)
```

```python
def delete_double_real_files(dataset_path="/mnt/ONCOBOX/geneplus/workspace/chenly/dataset"):
    # æ„å»ºåŒ¹é…æ¨¡å¼ï¼šåŒ¹é…æ‰€æœ‰ä»¥ double_real å¼€å¤´çš„æ–‡ä»¶
    img_pattern = os.path.join(dataset_path, "images/train/double_real_*.png")
    txt_pattern = os.path.join(dataset_path, "labels/train/double_real_*.txt")
    
    # è·å–æ–‡ä»¶åˆ—è¡¨
    img_files = glob.glob(img_pattern)
    txt_files = glob.glob(txt_pattern)
    
    all_files = img_files + txt_files
    
    if not all_files:
        print(" color=#cccccc>æœªæ‰¾åˆ°ä»»ä½• double_real ç›¸å…³æ–‡ä»¶ï¼Œè¯·æ£€æŸ¥è·¯å¾„ã€‚")
        return

    print(f"æ‰¾åˆ° {len(img_files)} å¼ å›¾ç‰‡å’Œ {len(txt_files)} ä¸ªæ ‡æ³¨æ–‡ä»¶ï¼Œå‡†å¤‡åˆ é™¤...")
    
    count = 0
    for f in all_files:
        try:
            os.remove(f)
            count += 1
        except Exception as e:
            print(f"âŒ åˆ é™¤å¤±è´¥ {f}: {e}")
            
    print(f"ğŸ§¹ æ¸…ç†å®Œæ¯•ï¼å…±åˆ é™¤ {count} ä¸ªæ–‡ä»¶ã€‚")

# æ‰§è¡Œåˆ é™¤
delete_double_real_files()
```
